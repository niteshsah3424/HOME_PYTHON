```python
#if...else
#1.wap to chek whether number is even is odd

num=int(input("enter a number"))
if num%2==0:
  print(num,"is the even number ")

else:
  print(num,"is not odd number")
```

```python
#2.wap to check number is divisible by 9 or not
num=int(input("enter the number"))
if num%9==0:
  print(num,"is divisible by nine")

else:
  print(num,"is not divisible by nine")
```

```python
#3.wap to check given string is pallindrom or not

str=input("enter a string")

if str==str[::-1]:
  print(str,"is a pallindrome string")

else:
  print(str,"is not a pallindrome string")


```
