```python
#1.wap to print kiit is a good college upto 10

i=0
while i <=10:
  print("kiit is a good college")

  i+=1
```

```python
#2.wap to print even number upto 100

i = 1
while i <= 100:
    if i % 2 == 0:
        print(i, end=" ")
    i += 1

```

```python
#3.wap to print multiplication table by using user table
num=int(input("enter a table number"))
i=1
n=10
while i<=n:
  print(f'{num}*{i}={num*i}')
  i+=1

```

```python
#4.wap to find the sum of natural number
num=int(input("enter a number"))

i=1
sum=0
while i<=num:
  sum+=i
  i+=1
print("sum of natural number of is ",sum)


```

```python
#5.wap to find the factoral of given num
num=int(input("enter a mumber"))

i=1
fact=1
while i<=num:
  fact*=i
  i+=1

print(f'the factorial of{num} is {fact}')
```

```python
#6.wap to get the the following output

s="we are kiit students we are the best developers"
output="we_are_kiit_students_we_are_the_best_developers"

out=""
i=0
while i<len(s):
  if s[i]==' ':
    out+='_'

  else:
    out+=s[i]

  i+=1

print(out)



```

```python
#7.wap to extract lower case character from the given str

s="Hello How ARE yoU?"
output="elloowyo"

out=""
i=0
while i<len(s):
  if 'a'<=s[i]<='z':
    out+=s[i]
  i+=1

print(out)

```

```python
#8.wap to extract special case character from the given str

s="kitne #$%^& aadmi 12345678 they $%^&() sardar 6 aadmi %^&*() they"
output="#$%& $%^&()  %^&*()"

out=""
i=0
while i<len(s):
  if not ('a'<=s[i]<='z'or 'A'<=s[i]<='Z' or '0'<=s[i]<='9'):

    out+=s[i]
  i+=1

print(out)
```

```python
#9. wap to extract only integer from a given list

l=[10,345,5.6, True,567,'kiit',[1,2,3,4,5]]
output=[10,234,567]

out=[]
i=0
while i< len(l):
  if type(l[i])==int:
      out.append(l[i])

  i+=1

print(out)






```
